```python
import re
import markdownify

def convert_to_md(html_content):
    temp_content = markdownify.markdownify(html_content)
    temp_content = re.sub(r'\n\s*\n', '\n\n', temp_content)
    temp_content = re.sub(r'Document generated by Confluence on .*\n', '', temp_content)
    temp_content = temp_content.replace('[Atlassian](https://www.atlassian.com/)', '')
    new_content = re.sub(r'\n{2,}', '\n\n', temp_content)
    return new_content
# print(new_content)
# print(convert_to_md(content))
```

```python
output_directory = r'/Users/wangjialong/Documents/AI/SAIC - DCSGlobalVehicle/全球整车系统设计/GLBDCSVE_OUTPUT'
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

for index, item in enumerate(files, start=1):
    full_path = os.path.join(file_directory, item)
    new_item_name = item.replace('.html', '.md')
    if os.path.isfile(full_path):
        print(f"处理文件 {index} |'{full_path}'")
        try:
            with open(full_path, 'r') as file:
                content = file.read()
                md_result = convert_to_md(content);
                with open(os.path.join(output_directory, new_item_name), 'w') as wfile:
                    wfile.write(md_result)                    
        except Exception as e:
            print(f"Failed! {index} | {full_path} | {e}")
        
```
